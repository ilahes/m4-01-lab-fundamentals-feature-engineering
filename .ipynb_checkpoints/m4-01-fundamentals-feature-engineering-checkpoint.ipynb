### Task 1: Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = sns.load_dataset("titanic")

# 1. First rows & shape
print("First 5 rows:")
print(titanic.head())

print("\nShape of dataset:")
print(titanic.shape)

# 2. Data types
print("\nData info:")
titanic.info()

# 3. Missing values
print("\nMissing values (count):")
print(titanic.isnull().sum())

print("\nMissing values (%):")
print((titanic.isnull().sum() / len(titanic)) * 100)

# 4. Target (survived) balance
print("\nSurvival counts:")
print(titanic['survived'].value_counts())

print("\nSurvival rate:")
print(titanic['survived'].mean())

# 5. Visualizations

# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(titanic['age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Survival by sex
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='survived', data=titanic)
plt.title("Survival Rate by Sex")
plt.show()
```

### Task 1 Summary

The Titanic dataset contains 891 rows and 15 columns, including both numerical and categorical features. Numerical features include variables such as age, fare, and number of siblings/spouses (sibsp), while categorical features include sex, embarked, and class.

Several columns contain missing values. The age column has about 20% missing data, while the deck column has more than 77% missing values, making it largely unusable. A small number of missing values are also present in embarked and embark_town.

The target variable (survived) is somewhat imbalanced. Around 38% of passengers survived, while 62% did not.

The age distribution shows that most passengers were between 20 and 40 years old, with fewer elderly passengers.

From the survival analysis, there is a clear difference between genders. Female passengers had a much higher survival rate (around 75%) compared to male passengers (around 19%). This suggests that gender was a strong factor influencing survival.

```python
# Task 2: Feature Engineering

# Titanic dataset-in kopyasını götürək ki, original data dəyişməsin
df = titanic.copy()

# 1. Handle missing values
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# deck column çox missing olduğu üçün drop edirik
df = df.drop(columns=['deck'])

# 2. Encode categorical features

# sex: male -> 1, female -> 0
df['sex'] = df['sex'].map({'female': 0, 'male': 1})

# embarked üçün one-hot encoding
df = pd.get_dummies(df, columns=['embarked'], drop_first=False)

# bool sütunları int-ə çevirək
bool_cols = df.select_dtypes(include=['bool']).columns
df[bool_cols] = df[bool_cols].astype(int)

# 3. Create new features
df['family_size'] = df['sibsp'] + df['parch'] + 1
df['is_alone'] = (df['family_size'] == 1).astype(int)

# 4. Drop non-numeric / unnecessary columns
df = df.drop(columns=['alive', 'who', 'adult_male', 'embark_town', 'class', 'alone'])

# Əgər category/object tipli əlavə sütun qalıbsa onları da çıxaraq
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns
print("Non-numeric columns before final cleanup:", list(non_numeric_cols))

df = df.drop(columns=non_numeric_cols)

# 5. Display final dataframe and confirm all columns are numeric
print("Final DataFrame:")
print(df.head())

print("\nShape after feature engineering:")
print(df.shape)

print("\nData types:")
print(df.dtypes)

print("\nAny non-numeric columns left?")
print(df.select_dtypes(exclude=[np.number]).columns.tolist())
```

### Task 2 Summary

In this step, missing values were handled by filling the age column with the median and the embarked column with the most frequent value. The deck column was removed due to a high percentage of missing values.

Categorical variables were converted into numeric format. The sex column was label encoded, and the embarked column was transformed using one-hot encoding.

Two new features were created: family_size, representing the total number of family members traveling with the passenger, and is_alone, indicating whether the passenger was traveling alone.

Unnecessary and non-numeric columns were removed, resulting in a clean dataset with only numeric features. This prepared the data for further analysis and machine learning modeling.

### Task 3: Feature Scaling & Selection

```python
# Task 3: Feature Scaling & Selection

from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Copy dataset
df_scaled = df.copy()

# Separate features (X) and target (y)
X = df_scaled.drop(columns=['survived'])
y = df_scaled['survived']

# -----------------------------
# 1. Compare Scalers
# -----------------------------

# Apply StandardScaler
standard_scaler = StandardScaler()
X_standard = pd.DataFrame(
    standard_scaler.fit_transform(X),
    columns=X.columns
)

# Apply MinMaxScaler
minmax_scaler = MinMaxScaler()
X_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(X),
    columns=X.columns
)

# Plot distributions for age and fare
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(X['age'], kde=True)
plt.title("Original Age")

plt.subplot(1,2,2)
sns.histplot(X_standard['age'], kde=True)
plt.title("Standard Scaled Age")

plt.show()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(X['fare'], kde=True)
plt.title("Original Fare")

plt.subplot(1,2,2)
sns.histplot(X_minmax['fare'], kde=True)
plt.title("MinMax Scaled Fare")

plt.show()

# -----------------------------
# 2. Correlation Analysis
# -----------------------------

corr_matrix = df_scaled.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Correlation with target
corr_with_target = corr_matrix['survived'].sort_values(ascending=False)
print("\nCorrelation with survived:")
print(corr_with_target)

# -----------------------------
# 3. Feature Selection
# -----------------------------

# Select features with |correlation| >= 0.1
selected_features = corr_with_target[abs(corr_with_target) >= 0.1].index.tolist()

# Remove target itself
selected_features.remove('survived')

print("\nSelected Features (|corr| >= 0.1):")
print(selected_features)
```

### Task 3 Summary

In this step, two different scaling techniques were applied: StandardScaler and MinMaxScaler. StandardScaler transforms features to have a mean of 0 and a standard deviation of 1, while MinMaxScaler scales features to a range between 0 and 1. The distributions show that StandardScaler centers the data, while MinMaxScaler compresses values into a fixed range.

The correlation matrix was used to examine relationships between features and the target variable. The strongest positive correlation with survival was observed for the sex feature (being female), while features such as pclass and is_alone showed negative correlations.

Based on the correlation analysis, features with an absolute correlation greater than or equal to 0.1 were selected. These features are considered the most relevant for predicting survival and will be useful for building machine learning models.

### Task 4: Dimensionality Reduction

```python
# Task 4: PCA (Dimensionality Reduction)

from sklearn.decomposition import PCA

# 1. Standardize all features (PCA üçün vacibdir)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 2. PCA with all components
# -----------------------------
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_scaled)

explained_variance = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Plot explained variance
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.bar(range(len(explained_variance)), explained_variance)
plt.title("Explained Variance Ratio")

plt.subplot(1,2,2)
plt.plot(range(len(cumulative_variance)), cumulative_variance, marker='o')
plt.axhline(y=0.8, linestyle='--')  # 80% xətti
plt.title("Cumulative Explained Variance")

plt.show()

# neçə component lazımdır?
n_components_80 = np.argmax(cumulative_variance >= 0.8) + 1
print(f"Number of components to reach 80% variance: {n_components_80}")

# -----------------------------
# 3. PCA with 2 components
# -----------------------------
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled)

# Scatter plot
plt.figure(figsize=(6,5))
sns.scatterplot(x=X_pca_2[:,0], y=X_pca_2[:,1], hue=y)
plt.title("PCA (2 Components)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

### Task 4 Summary

In this step, PCA was applied to reduce the dimensionality of the dataset. First, all features were standardized using StandardScaler, as PCA requires features to be on the same scale.

The explained variance ratio showed that the first few principal components capture a large portion of the total variance. The cumulative explained variance indicated that around 5 components are enough to retain at least 80% of the information.

When the data was projected into two principal components, the scatter plot showed partial separation between survivors and non-survivors. However, the classes were not clearly separable, indicating that PCA alone does not fully distinguish between them.

Overall, PCA is useful for dimensionality reduction and visualization, but the original feature space may still be more effective for classification tasks in this dataset.
